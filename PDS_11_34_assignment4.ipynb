```python

import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/PulockDas/pd-12-resources/refs/heads/master/titanic.csv"
df = pd.read_csv(url)
# Filter passengers who were alive (assuming 'Survived' column exists, where 1 means alive)
df['Number of Relatives'] = df['SibSp'] + df['Parch']
alive_passengers = df[df['Survived'] == 1]

plt.figure(figsize=(8, 6))
plt.scatter(alive_passengers['Number of Relatives'], alive_passengers['Fare'], alpha=0.5)
plt.xlabel('Number of Relatives')
plt.ylabel('Fare')
plt.title('Scatter Plot of Fare vs Number of Relatives (Alive Passengers)')
plt.grid(True)
plt.show()
```
